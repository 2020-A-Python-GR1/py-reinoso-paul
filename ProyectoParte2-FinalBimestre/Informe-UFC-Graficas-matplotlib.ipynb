### `3.- Cual es el promedio de golpes que recibira un luchador al año, en: HEAD,BODY,LEG`

```python
#Promedio de Golpes acertados en combates a la cabeza cuerpo clinch y ground
serie_head=df_limpio['R_HEAD']
serie_body=df_limpio['R_BODY']
serie_chinch=df_limpio['R_DISTANCE']
serie_ground=df_limpio['R_LEG']

serie_head_dos=df_limpio['B_HEAD']
serie_body_dos=df_limpio['B_BODY']
serie_chinch_dos=df_limpio['B_DISTANCE']
serie_ground_dos=df_limpio['B_LEG']

def calculo_promedio_serie(serie_seg):
    golpes_caveza = []
    for valor_serie in serie_seg:
        valores_separados = valor_serie.split('of')
        golpes_caveza.append(int(valores_separados[0]))    
    golpes_c = pd.Series(golpes_caveza)    

    suma = 0
    numero_valores = 0
    for valor_serie in golpes_c:    
        valor = int(valor_serie)      # valor a entero
        numero_valores = numero_valores + 1 #sumanmos todos los valores para saber por cuanto dividir
        suma = suma + valor                 # sumamos los valores de la columna        
    promedio = suma / numero_valores #calculamos el promedio
    return int(promedio)

def arreglo_en_promedio(serie):
    golpes_c = [] #rangos se estiman con un valor estimado de los combates al año de un luchador
    golpes_c.append(calculo_promedio_serie(serie.iloc[:156:]))
    golpes_c.append(calculo_promedio_serie(serie.iloc[156:312:]))
    golpes_c.append(calculo_promedio_serie(serie.iloc[312:468:]))
    golpes_c.append(calculo_promedio_serie(serie.iloc[468:780:]))
    golpes_c.append(calculo_promedio_serie(serie.iloc[780:936:]))
    golpes_c.append(calculo_promedio_serie(serie.iloc[936:1092:]))

    return golpes_c

def suma_peleadores_golpes_conbate(serie_a,serie_b):
    nueva = []
    for i in range(len(arreglo_en_promedio(serie_a))):
        nueva.append((arreglo_en_promedio(serie_a))[i]+(arreglo_en_promedio(serie_b))[i])
    return nueva
          
head=suma_peleadores_golpes_conbate(serie_head,serie_head_dos)
body=suma_peleadores_golpes_conbate(serie_body,serie_body_dos)
arm=suma_peleadores_golpes_conbate(serie_chinch,serie_chinch_dos)
leg=suma_peleadores_golpes_conbate(serie_ground,serie_ground)
```

```python
import matplotlib.ticker as ticker
import numpy as np

year = [2019, 2018, 2017, 2016, 2015, 2014]
population_by_continent = {
    'HEAD': head,
    'BODY': body,    
    'ARM':arm,
    'LEG': leg
    
}

fig, ax = plt.subplots()
fig.set_size_inches(12,6)
ax.grid(True)
ax.stackplot(year, population_by_continent.values(),
             labels=population_by_continent.keys())
ax.legend(loc='upper left')
ax.set_title('Promedio de golpes que recibira un luchador al año')
ax.set_xlabel('Años')
ax.set_ylabel('Número de golpes')
plt.yticks(range(0,150,5))

plt.show()
```
